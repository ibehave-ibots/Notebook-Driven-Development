```python
df = pd.read_csv("https://uni-bonn.sciebo.de/s/QcC6e3LXiNlNl0X/download")
df = df.drop(columns=["Unnamed: 0"])
df = df.dropna()
df = df[df['active_trials'] == True]

continuous_variables = ['gocue_time', 'response_time', 'feedback_time']
categorical_variables = ['response_type', 'feedback_type', 'mouse']

mice_names = list(df['mouse'].unique())
continuous_variables = ['gocue_time', 'response_time', 'feedback_time']

mouse_selector = pn.widgets.Select(name='Select Mouse', options=mice_names)

def update_dashboard(sm):
    # Summary statistics of continuous variables
    summary_stats = df[df['mouse'] == sm][continuous_variables].describe().T
    
    # Response type bar chart
    response_type_map = {-1: 'Left turn', 0: 'No turn', 1: 'Right turn'}
    filtered_df = df[df['mouse'] == sm].copy()
    filtered_df['response_type'] = filtered_df['response_type'].map(response_type_map)
    rt = filtered_df['response_type'].value_counts().reset_index()
    rt.columns = ['response_type', 'count']
    rt_plot = rt.hvplot.bar(x='response_type', y='count', rot=45, xlabel='Wheel Turn Direction', ylabel='Number of Trials')
    
    # Feedback type bar chart
    feedback_type_map = {-1: 'White Noise', 1: 'Water'}
    filtered_df['feedback_type'] = filtered_df['feedback_type'].map(feedback_type_map)
    fb = filtered_df['feedback_type'].value_counts().reset_index()
    fb.columns = ['feedback_type', 'count']
    fb_plot = fb.hvplot.bar(x='feedback_type', y='count', rot=45, xlabel='Feedback', ylabel='Number of Trials')
    
    return pn.Column(
        pn.Row(summary_stats),
        pn.Row(rt_plot, fb_plot)
    )

dashboard = pn.bind(update_dashboard, sm=mouse_selector)
layout = pn.Column(mouse_selector, dashboard)
layout.servable()
```
